### K-Means Clustering and Evaluation

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 2)
```

```python
CD.shape
```

```python
CD = CD.drop('Unnamed: 0', axis = 1)
```

```python
kmeans.fit(CD.drop('Private', axis = 1))
```

```python
kmeans.cluster_centers_
```

Evaluation

```python
def converter(priv):
    if priv == 'Yes':
        return 1
    else:
        return 0
```

```python
CD['Cluster'] = CD['Private'].apply(converter)
```

```python
CD.head(15)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print("\t\t\tEvaluation\n\nClassification Report\n")
print(classification_report(CD['Cluster'], kmeans.labels_))
print("\nConfusion Matrix\n")
print(confusion_matrix(CD['Cluster'], kmeans.labels_))
```
